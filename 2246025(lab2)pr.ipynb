```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(42)
```

parameters


```python
n = 100
learning_rate = 0.1
iterations = 500
noise  = 0.1
degree = 3
```

**Generate Data**

```python
x = np.linspace(0,1,n).reshape(-1,1)
t = np.sin(2 * np.pi * x) + np.random.normal(0,0.1,size = (n, 1))
```

Add bias

```python
X = np.hstack([x**i for i in range(degree + 1)])
w = np.random.randn(degree + 1,1)
```

Store loss for ploting

```python
loss_history = []
```

```python
loss_history = [] # Clear the loss history before starting the loop
for i in range(iterations):
  y_pred = X @ w
  error = y_pred - t  # Ensure error has the correct shape
  loss = (1 / (2 * n)) * np.sum(error ** 2)
  loss_history.append(loss)
  gradient = ( 1 /n) * X.T @ error
  w -= learning_rate * gradient
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
plt.scatter(x,t,label = 'Noisy data',color = 'blue')
plt.plot(x,X @ w,color = 'red',label = 'Linear Regression')
plt.xlabel ('x')
plt.ylabel('t')
plt.title('Linear Regression fit')
plt.legend()
plt.subplot(1,2,2)
plt.plot(range(iterations),loss_history,color = 'green')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.title('loss per iterations')
plt.tight_layout()
plt.show()
```

```python
n = 100
learning_rate = 0.1
iterations = 500
noise  = 0.1
degree = 9
learning_rates = [0.001,0.01,0.05,0.1,0.5,1.0]
np.random.seed(42)
```

```python
x = np.linspace(0,1,n).reshape(-1,1)
t = np.sin(2 * np.pi * x) + np.random.normal(0,0.1,size = (n, 1))
```

```python
plt.figure(figsize = (10,6))
for lr in learning_rates:
  X = np.hstack([x**i for i in range(degree + 1)])
  w = np.random.randn(degree + 1,1)
  loss_history = []
  for _ in range(iterations):
    y_pred = X @ w
    error = y_pred - t
    loss = (1/(2 * n)) * np.sum(error ** 2)
    loss_history.append(loss)
    gradinet = (1/n) * X.T @ error
    w -= lr * gradinet
  plt.plot(range(iterations),loss_history,label = f'lr = {lr}')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.title('loss per iterations for different learning rates')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Implement Stochastic Gradient Descent (SGD)
w_sgd = np.random.randn(degree + 1, 1)
loss_history_sgd = []
for i in range(iterations):
  # Randomly select a single data point for the gradient update
  random_index = np.random.randint(n)
  X_sgd = X[random_index].reshape(1, -1)
  t_sgd = t[random_index].reshape(1, -1)

  # Calculate the prediction and error for the selected data point
  y_pred_sgd = X_sgd @ w_sgd
  error_sgd = y_pred_sgd - t_sgd

  # Calculate the loss over the full dataset for monitoring (optional but good for comparison)
  y_pred_full = X @ w_sgd
  loss_sgd = (1 / (2 * n)) * np.sum((y_pred_full - t) ** 2)
  loss_history_sgd.append(loss_sgd)

  # Calculate the gradient based on the single data point
  gradient_sgd = X_sgd.T @ error_sgd # No division by n for a single data point

  # Update the weights
  w_sgd -= learning_rate * gradient_sgd

print("SGD training complete.")
```

```python
# Plot the loss history for both GD and SGD
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), loss_history, label='Gradient Descent (GD)')
plt.plot(range(iterations), loss_history_sgd, label='Stochastic Gradient Descent (SGD)')
plt.xlabel('iterations')
plt.ylabel('loss')
plt.title('Loss per Iteration: GD vs SGD')
plt.legend()
plt.grid(True)
plt.show()
```
